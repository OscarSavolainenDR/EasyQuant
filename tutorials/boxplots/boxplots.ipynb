# Per-channel boxplots for weight tensors

Here we demonstrate how one can use the per-output channel boxplot plotting API to do... what it says haha. First we will define a basic model.

```python
from quant_vis.boxplots import per_channel_boxplots
import torch

class Model(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.conv = torch.nn.Conv2d(3, 3, (3,3))

model = Model()


```

Then we will plot the per-channel boxplots. We set the `folder_path` and `filename` to be `None`, which emans we don't save the plots.

```python
per_channel_boxplots(
    weight_tensor=model.conv.weight,
    folder_path=None,
    filename=None,
    title='The last weight tensor'
)
```

Here we save the plots by specifying the path and filename. Both need to not be `None` for the plots to be saved.
